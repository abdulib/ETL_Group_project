Import the election data.

```python
import csv
import pandas as pd
```

```python
electionfile = "../data/clean/1976-2018-senate.csv"
election_df = pd.read_csv(electionfile)
election_df.head()
```

Slice and dice the data to retain only  what we really need.,

```python
xyz = ["year","state","state_po","candidate","party","candidatevotes","totalvotes"]
election_df = election_df[xyz]
election_df.head()
```

```python
election_df = election_df.loc[election_df["year"] > 2005, :]
# election_df = election_df.loc[election_df["stage"] == "gen", :]
```

```python
election_df.head()
```

Identify the election winners

```python
election_df = election_df.sort_values(["year","state","candidatevotes"], ascending=[True, True, False])
election_df.head()
election_df.info()
```
