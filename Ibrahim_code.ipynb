```python
date_introduced = str(bills[0]('span')[3].text)
date_introduced = date_introduced[date_introduced.find("(")+1:date_introduced.find(")")].split()[1]
date_introduced
```

```python
date_introduced_list = []
bill_num_list = []
bill_name_list = []
senator_details_list = []
last_name_list = []
first_name_list = []
cosponsors_list= []
party_list = []
state_list = []

for bill in bills:
    bill_num = bill('a')[0].text
    senator_details = bill('a')[1].text
    last_name = senator_details.split()[1]
    first_name = senator_details.split()[2]
    cosponsors = bill('a')[2].text
    bill_name = bill('span')[2].text
    date_introduced = str(bill('span')[3].text)
    date_introduced = date_introduced[date_introduced.find("(")+1:date_introduced.find(")")].split()[1]
    party = senator_details.split()[-1][1]
    state = senator_details.split()[-1][3:5]
    
    

    print('------------------')
    print(date_introduced)
    print(bill_num)
    print(bill_name)
    print(senator_details)
    print(last_name)
    print(first_name)
    print(cosponsors)
    print(party)
    print(state)

    date_introduced_list.append(date_introduced)
    bill_num_list.append(bill_num)
    bill_name_list.append(bill_name)
    senator_details_list.append(senator_details)
    last_name_list.append(last_name)
    first_name_list.append(first_name)
    party_list.append(party)
    state_list.append(state)
    cosponsors_list.append(cosponsors)
    

    
```

```python
congress_bills_on_debt_data = pd.DataFrame({
    'Date Introduced': date_introduced_list,
    'Bill number': bill_num_list,
    'Bill Name': bill_name_list,
    'Senator Details': senator_details_list,
    'Last Name': last_name_list,
    'First Name': first_name_list,
    'Party': party_list,
    'State': state_list,
    'Number of Cosponsors': cosponsors_list
})
```
